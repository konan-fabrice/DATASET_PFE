```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture

# Charger et préparer les données (utiliser le DataFrame df_merged)
# Assurez-vous que df_merged est déjà filtré comme indiqué dans le code précédent

# Sélectionner les caractéristiques pour la segmentation
features = df_merged[['QUANTITE']]  # Ajouter d'autres caractéristiques si nécessaire

# Normaliser les données
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# 1. K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
df_merged['KMeans_Cluster'] = kmeans.fit_predict(features_scaled)

# 2. DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
df_merged['DBSCAN_Cluster'] = dbscan.fit_predict(features_scaled)

# 3. Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=3, random_state=42)
df_merged['GMM_Cluster'] = gmm.fit_predict(features_scaled)

# Visualisation des résultats
plt.figure(figsize=(16, 12))

# K-Means
plt.subplot(3, 1, 1)
sns.scatterplot(data=df_merged, x='QUANTITE', y='QUANTITE', hue='KMeans_Cluster', palette='viridis', s=100)
plt.title('Segmentation avec K-Means')

# DBSCAN
plt.subplot(3, 1, 2)
sns.scatterplot(data=df_merged, x='QUANTITE', y='QUANTITE', hue='DBSCAN_Cluster', palette='coolwarm', s=100)
plt.title('Segmentation avec DBSCAN')

# Gaussian Mixture Model (GMM)
plt.subplot(3, 1, 3)
sns.scatterplot(data=df_merged, x='QUANTITE', y='QUANTITE', hue='GMM_Cluster', palette='plasma', s=100)
plt.title('Segmentation avec Gaussian Mixture Model (GMM)')

plt.tight_layout()
plt.show()

# Affichage des résultats
print(df_merged[['CLIENT', 'QUANTITE', 'KMeans_Cluster', 'DBSCAN_Cluster', 'GMM_Cluster']].head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score

# Charger les données fusionnées
df_merged = pd.read_csv("chemin/vers/ton/fichier.csv")

# Supposons que nous voulons prédire la 'COMMUNE' en fonction de 'QUANTITE' et 'VILLE'
X = df_merged[['QUANTITE']]  # Caractéristiques
y = df_merged['COMMUNE']      # Cible

# Séparer les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créer l'arbre de décision
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluation du modèle
print("Rapport de Classification :")
print(classification_report(y_test, y_pred))
print("Précision :", accuracy_score(y_test, y_pred))

# Visualiser l'arbre de décision
plt.figure(figsize=(12, 6))
plot_tree(model, filled=True, feature_names=['QUANTITE'], class_names=model.classes_)
plt.title('Arbre de Décision')
plt.show()
```

```python
# Utiliser l'Elbow Method pour déterminer le nombre optimal de clusters

wcssmin = []  # Within-cluster sum of squares

for i in range(1, 30):  # Essayer pour k de 1 à 20 clusters
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(x)
    wcssmin.append(kmeans.inertia_)

# Afficher la courbe du "Elbow"
plt.plot(range(1, 30), wcssmin)  # Pas de 1 pour les clusters

# Personnaliser l'axe des x avec un pas de 1
plt.xticks(range(1, 30))  # Ajuste les ticks de l'axe des x de 1 à 20 avec un pas de 1

plt.title('Méthode du Coudé (Elbow Method)')
plt.xlabel('Nombre de clusters')
plt.ylabel('WCSS')
plt.show()
```
